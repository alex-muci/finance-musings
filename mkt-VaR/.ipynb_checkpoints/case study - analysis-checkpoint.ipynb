## Task 2:  1y risk estimate

1. First - for the equities desk - we calculate the tracking error and explain why we choose the zero-mean standard deviation definition.

2. Then we look at the risk estimate over 1y horizon of the equity desk, CIO and CEO allocations.
We focus on two standard (related) measures: 
 - VaR, i.e. the loss level that will not be exceeded with a certain confidence level during a certain holding period, and
 - CVaR or Expected Shortfall, i.e. the average amount that is lost over a certain holding period assuming the loss is greater than the related VaR amount.

Both estimate are a function of (i) time horizon (here, 1 year) and confidence level (here, we assume - following insurance/pension standard, e.g. Moody's - a conf. level of 95%).

We will also look at the whole distribution over 1y and finally to another, more longer-term measure, to make more self-evident the usual trade-off between goals (e.g. being fully funded in 10-15 years) and related risk. 

### Tracking error from equities desk

Some data wrangling:

```python
trading_days_1y = 252  # days in 1y

# get 1y rolling FTSE returns (daily close-to-close)
ftse_close_1yrollingRtrns = ftse_close/ftse_close.shift(trading_days_1y) - 1
f = ftse_close_1yrollingRtrns.rename("1y_FTSE_rtrns")  # rename before concat

# equity desk rthrns == simply 1.1 yearly rtns of FTSE
eq = (ftse_close_1yrollingRtrns *1.1).rename("1y_equity_rtrns")

ftse = pd.concat([ftse_close.rename("FTSE_lvl"), f, eq], axis=1).dropna()
ftse.tail(3)
```

Let's use 3 definitions of tracking errors plus 1 that I find intuitive:

```python
# 1. the simple difference between mgr and bmk sometimes called tracking error
ftse['rtrn_diff'] = ftse['1y_equity_rtrns'] - ftse['1y_FTSE_rtrns']

roll_per = 21 * 6    # 6 months

# 2a. a more common definition of tracking error (st dev of rtrn diff)
ftse['tracking_error'] = ftse['rtrn_diff'].rolling(roll_per).std()

# 2b simple sqrt avg of square of rtrn diff (i.e. 2a but with no mean)
ftse['tracking_error_0mean'] = np.sqrt(np.square(ftse['rtrn_diff']).rolling(roll_per).mean())

# 3. more intuitive definition: rolling median of abs(rtrn_diff)
ftse['intuitive_tracking_error'] = np.abs(ftse['rtrn_diff']).rolling(roll_per).median()

TE  = ftse[['rtrn_diff', 
            'tracking_error', 'tracking_error_0mean',
            'intuitive_tracking_error']].dropna() # or use pd.concat for series, as above ...

# multiplots (OO style)
fig , (ax1 , ax2) = plt.subplots(nrows=1, ncols=2, 
                               sharey =True, figsize =(9 ,6)) # unpacking
ax1.plot(TE.index, TE['rtrn_diff'])
ax1.set_title("1y rolling rtrns diff")

ax2.plot(TE.index, TE[['intuitive_tracking_error']], color='orange')
ax2.set_title("Tracking errors (my intuitive version)")
fig.tight_layout()

# focusing on last three
TE.drop('rtrn_diff', axis=1).plot(figsize =(8 ,6))
```

We find out that one should try and avoid definition [2a] (i.e. standard deviation of the return difference) and use either [2c] (which is what is often meant by pratictioners, althought sometimes erroneously called std dev of rtrn diff) or [3] above.

```python
print(f"Traditional tracking error (last month mean): {np.mean(TE['tracking_error_0mean'].values[-21:])*100:.{3}}%")
print(f"Intuitive tracking error (last month mean): {np.mean(TE['intuitive_tracking_error'].values[-21:])*100:.{3}}%")
    
print("\nTE historical distributions (since 1981)")
plt.figure(figsize=(8,6))
plt.subplot(1 ,2 ,1)
sns.distplot(TE['tracking_error_0mean'], kde=True)
plt.subplot(1 ,2 ,2)
sns.distplot(TE['intuitive_tracking_error'] ,kde=True)
fig.tight_layout()
```

In the light of the above, the best measure seems to be the traditional tracking error (LP2-norm) but with zero mean (because of its smoother, not bi-modal, distribution vs. LP1-norm):

$TE_t = \sqrt{\sum^n_{i=1} \frac{ (R^{asset}_{t-i} - R^{bmk}_{t-i})^2}{n}}$


### CEO vs. CIO vs. Desk

We may just focus on the equity allocation alone (main risk), but it makes sense here to model the whole asset side to differentiate between CEO and CIO (differences in leverage and LDI exposure - we assume that the leverage cost is 50bps over LDI returns).

#### 1. Historical distributions

```python
# Represent asset-side for the 3 allocations

lev_spread = 50 / 10000  #  50 basis points, assumed for simplicity over LDI

yearly_rtrns['CEO assets YoY'] =   0.70 * yearly_rtrns['FTSE YoY']  \
                                 + 0.50 * yearly_rtrns['RPY YoY']   \
                                 - 0.20 * (yearly_rtrns['RPY YoY'] + lev_spread) 

yearly_rtrns['CIO assets YoY'] =   0.65 * yearly_rtrns['FTSE YoY']  \
                                 + 0.45 * yearly_rtrns['RPY YoY']   \
                                 - 0.10 * (yearly_rtrns['RPY YoY'] + lev_spread) 

yearly_rtrns['Eq Desk'] = yearly_rtrns['FTSE YoY'] * 1.10
yearly_rtrns['Desk assets YoY'] =  0.65 * yearly_rtrns['Eq Desk']  \
                                 + 0.45 * yearly_rtrns['RPY YoY']  \
                                 - 0.10 * (yearly_rtrns['RPY YoY'] + lev_spread)
```

```python
display(yearly_rtrns.tail(2))
yearly_rtrns.drop('Eq Desk', axis=1, inplace=True)
# yearly_rtrns.drop('rolling correl', axis=1, inplace=True)
display(yearly_rtrns.info())
```

```python
# print results
print("               HISTORICAL RISK STATISTICS")
list_of_impl=['CEO assets YoY', 'CIO assets YoY', 'Desk assets YoY']
#
for i in list_of_impl:
    
    print("\n{}" .format(i))    
    rtrns_array = yearly_rtrns[i].values
    
    _var =  np.percentile(rtrns_array, 100-95)                  # 95% VaR
    _Cvar = rtrns_array[rtrns_array<= _var].mean()            # 95% Expected shortfall or conditional VaR
    _N_var = (rtrns_array.mean() - 1.645 * rtrns_array.std())  # 95% 'Normal' VaR
    
    print("95% VaR: {}% (while 95% Normal VaR is: {}%) and 95% CVar (or ES): {}% " \
          .format(round(-_var*100,1), round(-_N_var*100,1), round(-_Cvar*100,1)))
```

From above we note that:
- **Normal VaR** (i.e. a VaR calculated assuming normally-distributed returns) **underestimates** the losses over 1y horizon (the figure clearly shows a fat left tail in the distribution). We also calculated the Expected Shortfall (a more 'coherent' risk measure).
- the Equity desk is adding unwanted risk (vs CEO and even CEO) - our intution is confirmed by the data.

#### 2. Responsiveness and stability of risk measures

Our calculations above are based on the fully available history (30y+), which obviously gives us a **stable** risk estimation. But we also want our estimates to be more indicative of current market conditions (i.e. **responsive**) in order to guide us in varying our allocation overtime. In other words, we would like to assess risk using *similar* data to our conditions today. The simplest assumption, in this regard, is that the recent past should weight more than the distant one. Other similarity concepts may be based on selecting only periods that display similar volatilities or closer Euclidean distances (e.g. KKN) or return-correlations (e.g. online machine learning like CORN). 

Of course stability and responsiveness are **opposite** features and the trade-off between the two is really down to their use and subjective preferences, e.g. a board/CEO would tend to weigh more the former, whilst a trading desk would definitely  prefer the latter. The whole idea can be recast in the desire to filter 'signal' from 'noise' and, as such, it is a common trade-off in finance (model calibration, trading signal, etc.) and other fields.

In this section, in order to make our VaR more relevant to current market conditions, we limit ourselves to the `last 3y of data`. Why 3 years? We are assuming that our estimates are for `CEO and board consumption` and that they tend to think in 3y horizons (similarly to the statutory funding reviews). Were the estimates for the CIO or the trading desk I would have used 1y lookback period and the more granular daily data of the FTSE only.

We will be using the `monthly (rather than the yearly) returns` of the assets (for all 3 possible configurations: CEO, CIO, Eq Desk) to bootstrap 1y returns and simulate our chosen risk statistics (1y 95% VaR and CVaR).

With this modelling choice, we are keeping the correlation structure between equity/LDI/leverage intact but at the expense of assuming independent monthly returns (no auto-correlations). Another choice would have been to impose more 'structure', e.g. assuming a model for the evolution of prices/returns (e.g. geometric Brownian motion) or, more simply, we could have use yearly data on the more granular FTSE data to simulate 1y returns (sampling 1y returns with replacement) and ignoring LDI/Leverage.

```python
# Represent asset-side for the 3 allocations
lev_spread = (50 / 10000) / 12. #  50 basis points per annum, assumed for simplicity over LDI

monthly_rtrns['CEO assets MoM'] =  0.70 * monthly_rtrns['FTSE MoM']  \
                                 + 0.50 * monthly_rtrns['RPY MoM']   \
                                 - 0.20 * (monthly_rtrns['RPY MoM'] + lev_spread)  # NB: Hp lev_spread scaled above

monthly_rtrns['CIO assets MoM'] =  0.65 * monthly_rtrns['FTSE MoM']  \
                                 + 0.45 * monthly_rtrns['RPY MoM']   \
                                 - 0.10 * (monthly_rtrns['RPY MoM'] + lev_spread) 

monthly_rtrns['Eq Desk'] = monthly_rtrns['FTSE MoM'] * (1 + 1.2/12.)  # NB: Hp returns scaling
monthly_rtrns['Desk assets MoM'] =  0.65 * monthly_rtrns['Eq Desk']  \
                                  + 0.45 * monthly_rtrns['RPY MoM']   \
                                  - 0.10 * (monthly_rtrns['RPY MoM'] + lev_spread) 

monthly_rtrns.drop('Eq Desk', axis=1, inplace=True)
```

```python
# let's limit the data to the past 3 years => 36 obs
monthly_rtrns_3y = monthly_rtrns.iloc[-36:]
```

```python
pd.DataFrame(monthly_rtrns_3y.mean()*12*100, columns=['annual avg rtrn (%)'])

print("        Summary Statistics")
mean = monthly_rtrns_3y.mean() * 12 * 100
stdev = monthly_rtrns_3y.std()* (12**0.5) * 100
IR = mean / stdev
pd.DataFrame({'annual avg rtrn (%)': mean, 
              'annual std dev (%)': stdev,
              'information ratio': IR} )
```

Build 1y path (bootstrapping), keep last returns and repeat:

```python
rnd.seed(101)
rtrns_per_year = 12  # 12 months
num_sims = 50000
final_rtrns = {}

start =  time.time()

for asset_loc in ['CEO assets MoM', 'CIO assets MoM', 'Desk assets MoM']:
    
    hist_returns = list(monthly_rtrns_3y[asset_loc].values)

    rtrns_r = [rnd.choices(hist_returns, k=rtrns_per_year) for _ in range(num_sims)]
    simulations = (1 + pd.DataFrame(rtrns_r).T).cumprod()

    final_rtrns[asset_loc] = simulations.iloc[-1].values - 1

end = time.time()
print("seconds: {}" .format(round(end - start, 2)))
```

Plot the results:

```python
sns.set_style("whitegrid")
fig1 = plt.figure(figsize=(6,4), dpi=100)  #, frameon=True, facecolor=None)

#plt.subplot(1, 3, 1)
sns.distplot(final_rtrns['CEO assets MoM'], 
             kde_kws={"color": "g", "lw": 3, "label": "CEO rtrn dist"},
             hist_kws={"histtype": "step", "linewidth": 0.3, "alpha": 0.5, "color": "g"})
#plt.subplot(1, 3, 2)
sns.distplot(final_rtrns['CIO assets MoM'],                         
             kde_kws={"color": "k", "lw": 2, 'ls': '-.', "label": "CIO rtrn dist"},
             hist_kws={"histtype": "step", "linewidth": 0.3, "alpha": 0.5, "color": "k"})
#plt.subplot(1, 3, 3)
sns.distplot(final_rtrns['Desk assets MoM'],              
             kde_kws={"color": "r", "lw": 2, 'ls': '--', "label": "Eq Desk rtrn dist"},
             hist_kws={"histtype": "step", "linewidth": 0.3, "alpha": 0.4, "color": "r"})
#
plt.xlabel("annual returns")
plt.ylabel("Probabilities (%)")
plt.legend(loc =2) # upper left corner
plt.grid(color ='b', alpha=0.3 , linestyle ='dashed', linewidth =0.5)
fig.patch.set_visible(True)

plt.tight_layout()
sns.despine()
fig1.savefig("fig2_SimulatedDist.png", dpi=200, transparent=True)
```

Calculate risk estimates (using previous code):

```python
print("\t\t\tSIMULATED RISK STATISTICS")
list_of_impl = ['CEO assets MoM', 'CIO assets MoM', 'Desk assets MoM']  
#
for i in list_of_impl:
    
    print("\n{}" .format(i))    
    rtrns_array = final_rtrns[i] #.values                                
    
    _var =  np.percentile(rtrns_array, 100-95)                  # 95% VaR
    _Cvar = rtrns_array[rtrns_array<= _var].mean()            # 95% Expected shortfall or conditional VaR
    _N_var = (rtrns_array.mean() - 1.645 * rtrns_array.std())  # 95% 'Normal' VaR
      
    print("95% VaR: {}% (while 95% Normal VaR is: {}%) and 95% CVar (or ES): {}% " \
          .format(round(-_var*100,1), round(-_N_var*100,1), round(-_Cvar*100,1)))
    
    mean = np.mean(rtrns_array)*100
    stdev = np.std(rtrns_array)*100
    print("annual return: {}, volatility {}, Info ratio {}"  .format(round(mean,1), round(stdev,1), round(mean/stdev,2) ))
```

The bootstrapped VaR is more benign vs. the 30y historical distributions above - something to be expected since we only use the last 3y of data, which were a relatively calm period. Indeed, VaR and ES display a less fat tail than experienced historically.

### Limitations

We have made a number of assumptions, like: 
- in our historical and bootstrapping distributions the RPI level was assigned at the end of each month (assuming no revisions) and coupled with the relevant FTSE level (mismatch);
- constantly rebalancing allocations across months;
- leverage cost (spread) was assumed constant (an heroic Hp);
- in the second bootstrapping simulation, returns are independent between months (forgoing volatility clustering, trending, etc.);
- completely ignored any liability-side considerations (demographic, regulatory risk, etc.)

But the biggest of all was that we assumed that the future (next 1y return distribution) is like the more recent past (i.e. that implied by the historical return distribution of the past 3y). 

The best way to overcome such limitations is to run a series of scenarios to enrich our states of the world. No crisis is similar to the past one, therefore being prepared and reactive is the best way to proceed. A simple deleveraging mechanism (e.g. vol cap/target on liquid assets, that should represent at least $\approx$ 70% of total) should help in most of such scenarios.


Below, we show a first attempt to extend our analysis by introducing liabilities vs. assets evolution overtime and time to reach full funding (starting from 80%).

```python
rnd.seed(101)
months =  np.array(range(12,306, 6))
years = list(months / 12)

num_sims = 1000

final_rtrns_combo, variance = {}, {}

start =  time.time()


for asset_loc in ['CEO assets MoM', 'CIO assets MoM', 'Desk assets MoM', 'RPY MoM']:
    final_rtrns_combo[asset_loc], variance[asset_loc] = [], []
    
    for month in months:
        
        month = 6  # sampling every 6 months: faster
        hist_returns = list(monthly_rtrns_3y[asset_loc].values)

        rtrns_r = [rnd.choices(hist_returns, k=month) for _ in range(num_sims)]
        simulations = (1 + pd.DataFrame(rtrns_r).T).cumprod()

        final_rtrns_combo[asset_loc].append(np.mean(simulations.iloc[-1].values - 1))
        variance[asset_loc].append(np.var(simulations.iloc[-1].values - 1))

end = time.time()
print("seconds: {}" .format(round(end - start, 2)))
```

```python
# get first values
years.insert(0,0)

portf_proj, vol_up, vol_down = {}, {}, {}
num_std_devs = 1.
for asset_loc in ['CEO assets MoM', 'CIO assets MoM', 'Desk assets MoM', 'RPY MoM']:
    
    port_start =  100. if asset_loc=='RPY MoM' else 80.
    
    new_rtrns = np.insert(final_rtrns_combo[asset_loc], 0, 0)
    portf_proj[asset_loc] = port_start * (1 + pd.DataFrame(new_rtrns)).cumprod()
    
    vvv = np.insert(variance[asset_loc], 0, 0)  #  or variance[asset_loc].insert(0, 0)
    var = pd.DataFrame(vvv)
    vol_up[asset_loc] = portf_proj[asset_loc] * (1 + num_std_devs * np.sqrt(var.cumsum()))
    vol_down[asset_loc] = portf_proj[asset_loc] * (1 - num_std_devs * np.sqrt(var.cumsum()))
```

Plot results

```python
sns.set_style("whitegrid")
fig1 = plt.figure(figsize=(6,4), dpi=100)
plt.plot(years, portf_proj['CEO assets MoM'], label='CEO assets growth', color='b', lw=1.5)
plt.plot(years, portf_proj['CIO assets MoM'], label='CIO assets growth', color='r', lw=1.5)
plt.plot(years, portf_proj['Desk assets MoM'], label='Desk assets growth', color='y', lw=1.5)
plt.plot(years, portf_proj['RPY MoM'], label='Liabilities growth', color='green', marker='o', ls='dashed', lw=2, markersize=3)

plt.ylabel("Assets, Liabilities (%)")
plt.xlabel("years")

plt.legend(loc =2) # upper left corner
plt.grid(color ='b', alpha=0.3 , linestyle ='dashed', linewidth =0.5)
fig.patch.set_visible(True)
plt.tight_layout()
sns.despine()

fig1.savefig("fig3_overtimeProfile.png", dpi=200, transparent=True)
```

## Task 3: CIO and eq. desk risk limits


The limits that may be applied to CIO and desk should reflect their different mandates, i.e. (i) CIO takes more contingent views on macro exposures (within pre-agreed risk limits) vs. CEO/board longer-term strategic allocations and (ii) equity desk transacts single instruments that implements equity exposure within stricter limits.

### CIO risk limits:
- *VaR* (or Expected Shortfall) at *portfolio* level to control for *correlation spikes* during stress periods (and hence explicit mandate to reduce exposure faster);
- counterparty (brokers) concentration risk;
- *differential VaR* for each *asset class* level vs. CIO implied ones (in our simple case it is mainly equity: $\Delta$ Var equity = VaR CIO equity - VaR CEO equity;
- *concentration* limits (incl. leverage) for each asset classes levels (e.g., equity vs. gilts (?) vs. corp bonds vs. foreign holdings);
- concentration limits within asset classes areas: credit (% IG vs. % Junk, CV01), rates (duration, PVO1), equity sectors (e.g. manufacturing, financials), implied foreign country exposure, alternatives (property, hedgies, commodities, ...).

### Equity Desk risk limits:
- tracking error limit to keep implementation in line with CIO mandated exposures;
- list of not allowed investments (e.g. 'too small' caps);
- granular concentration limits: single equity exposure limits (e.g. 4% for single shares/single futures), single external funds (15% diversified vs. 5% concentrated) and fund types (UCITs vs illiquid), single sectors (11 GICS), single group/geographic areas, aggregate limitation of bigger exposure (e.g. $<$ 40% exposures to greater than 5% single issuers)
- gross and net exposure limits (e.g. 150% and 100% of allocation, respectively);
- derivatives: all derivs (delta limits $\approx$ single exposure limits), options (e.g. vega $<$ 1M), OTC derivatives credit risk (4% max uncollateralised exposure - overall counterparty limit set by CIO);
- *liquidity* exposure (% cap < median turnover over 5 days) to be applied to shares representing, say, 85% of overall equity desk exposure;
- volatility cap control ($\ne$ targeting): reduce exposure when volatility too high (e.g. $1.5 (q_{0.75} - q_{0.25})$) to avoid value traps/defaults.


Of course similar limits should also be applied to non-equity investments (issuer concentrations and rating for corporate bond issuers, duration / key rate risk, etc.). 

Implicitly I have not imposed that the CIO should run some minimum risk exposures: too low risk exposures can be easily sanctioned in due time by the board (better to err on the side of being too cautious and then justify oneself), whilst taking excessive risk can quickly cause severe losses that are hard to recover. I have not considered governance issues like controlling shares and related.

Regarding the suggested liquidity limit for the equity desk: a pension scheme is, of course, in an almost perfect position to try and exploit any illiquidity premium, but here I am considering the equity desk exposure to be part of the 'liquid' bucket (limiting its illiquid holding to, say, 15% max of its allocation) so to be exploited by the CEO and, more likely, the CIO to quickly cut exposures in bear/stressed markets.

## Task 4: 1y risk pitfalls

The ultimate goal of a pension scheme is to be able to **pay benefits to members** as they fall due. This is a very long-term, **decade-long**, objective. Thus, *just* focusing on the 1y horizon misses a whole range of options available and cloud perspective; e.g. for a start, reaching full funding (starting from 80%) is not obviously advisable to do in just 12 months â€“ one needs a much longer-term plan in order to avoid taking excessive risks (and hence bearing a very high probability of ruin).

Of course, that does not mean that short-term focus must be disregarded. For example: (i) regulation demands that at least every 3y schemes need to carry a formal funding valuation or (ii) trading/risk mgt requires to periodically adjust exposures to ever changing reg/market conditions (strategic allocations and VaR targeting).

In short, focusing only on the 1y risk horizon misses the time profile of the trade-off between risks and returns (e.g. shorter-term VaR control vs. longer-term prob of reaching goals) and - in addition - it completely ignores some risks like longevity and inflation regime changes that materialise over longer cycles.
